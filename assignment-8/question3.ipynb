```python
# =============================================================================
# Assignment: 8
# Filename: question3.py
# Description: The Program that creates a GUI window
# Date: 03/03/2020
# Author: Tarini Dash
# =============================================================================

from graphics import *

# The display function    
def display(scoreA,scoreB,entry_A,entry_B):
    f_scoreA = 0
    f_scoreB = 0
    if(scoreA == 0):
        f_scoreA = 0
    elif(scoreA == 1):
        f_scoreA = 15
    elif(scoreA == 2):
        f_scoreA = 30
    elif(scoreA == 3):
        f_scoreA = 40
    if(scoreB == 0):
        f_scoreB = 0
    elif(scoreB == 1):
        f_scoreB = 15
    elif(scoreB == 2):
        f_scoreB = 30
    elif(scoreB == 3):
        f_scoreB = 40
    
    
    if(scoreA < 4 and scoreB < 4):
        entry_A.setText(f_scoreA)
        entry_B.setText(f_scoreB)
    else:
        if(scoreA > scoreB):
            f_scoreA = "Adv"
            f_scoreB = ""
        elif(scoreA < scoreB):
            f_scoreA = ""
            f_scoreB = "Adv"
        else:
            f_scoreA = "40"
            f_scoreB = "40"
        
        entry_A.setText(f_scoreA)
        entry_B.setText(f_scoreB)
        

    return f_scoreA, f_scoreB


def checkResult(f_scoreA,f_scoreB,listA,listB,win):
    if(len(listA) > 0 and f_scoreA == "Adv" and listA[-1] == "Adv"):
        Text(Point(200,70), "Player A won the game!").draw(win)
    if(len(listB) > 0 and f_scoreB == "Adv" and listB[-1] == "Adv"):
        Text(Point(200,70), "Player B won the game!").draw(win)
    listA.append(f_scoreA)
    listB.append(f_scoreB)


# function to check of the click occurred in a rectangle        
def is_button(click, rect):
    p1 = rect.getP1()
    p2 = rect.getP2()
    if (click.getX() > p1.getX()) and (click.getX() < p2.getX()):
        if (click.getY() > p1.getY()) and (click.getY() < p2.getY()):
            # the click is within the rectangle, return True
            return True

# The main function        
def main():
    win = GraphWin('Question-3', 400, 300)
    rect = Rectangle(Point(50,50), Point(350,250))
    rect.setOutline('black')
    rect.draw(win)

    rect_1 = Rectangle(Point(150,130), Point(190, 150))
    rect_1.setOutline('black')
    rect_1.setFill('grey')
    rect_1.draw(win)
    
    rect_2 = Rectangle(Point(300, 130), Point(340, 150))
    rect_2.setOutline('black')
    rect_2.setFill('grey')
    rect_2.draw(win)
    
    rect_3 = Rectangle(Point(180, 180), Point(240, 220))
    rect_3.setOutline('black')
    rect_3.setFill('red')
    rect_3.draw(win)
    
    Text(Point(100,105), "Score of player A:").draw(win)
    Text(Point(100,140), "Click here if player \n A wins a point:").draw(win)
    Text(Point(250,105), "Score of player B:").draw(win)
    Text(Point(250,140), "Click here if player \n B wins a point:").draw(win)
    Text(Point(210,200), "Quit").draw(win)
    
    scoreA = 0
    scoreB = 0
    listA = []
    listB = []
    entry_A = Entry(Point(170,105),5)
    entry_A.setFill('white')
    entry_A.setText(scoreA)
    entry_A.draw(win)
    
    entry_B = Entry(Point(320,105),5)
    entry_B.setFill('white')
    entry_B.setText(scoreB)
    entry_B.draw(win)
    
   
    while True:
        click = win.checkMouse()
        if click != None:
            if is_button(click,rect_1):
                scoreA += 1
                f_scoreA,f_scoreB = display(scoreA,scoreB,entry_A,entry_B)
                checkResult(f_scoreA,f_scoreB,listA,listB,win)
            elif is_button(click,rect_2):
                scoreB += 1
                f_scoreA,f_scoreB = display(scoreA,scoreB,entry_A,entry_B)
                checkResult(f_scoreA,f_scoreB,listA,listB,win)
            elif is_button(click,rect_3):
                break;
    
    win.close()
    
    
main()
```
