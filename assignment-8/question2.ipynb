```python
# =============================================================================
# Assignment: 8
# Filename: question2.py
# Description: The Program that creates a GUI window
# Date: 03/03/2020
# Author: Tarini Dash
# =============================================================================
from graphics import *

# function to check of the click occurred in a rectangle        
def is_button(click, rect):
    p1 = rect.getP1()
    p2 = rect.getP2()
    if (click.getX() > p1.getX()) and (click.getX() < p2.getX()):
        if (click.getY() > p1.getY()) and (click.getY() < p2.getY()):
            # the click is within the rectangle, return True
            return True

# The main function        
def main():
    win = GraphWin('Question-2', 400, 300)
    rect = Rectangle(Point(50,50), Point(350,250))
    rect.setOutline('black')
    rect.draw(win)

    rect_1 = Rectangle(Point(100, 160), Point(175, 220))
    rect_1.setOutline('black')
    rect_1.setFill('yellow')
    rect_1.draw(win)
    
    rect_2 = Rectangle(Point(225, 160), Point(300, 220))
    rect_2.setOutline('black')
    rect_2.setFill('red')
    rect_2.draw(win)
    Text(Point(137,190), "Calculate BMI").draw(win)
    Text(Point(265,190), "Quit").draw(win)
    Text(Point(200,70), "Your height in inches:").draw(win)
    Text(Point(195,100), "Your weight in pounds:").draw(win)
    Text(Point(222,130), "Your BMI is:").draw(win)
    
    entry_height = Entry(Point(280,70),5)
    entry_height.setFill('white')
    entry_height.draw(win)
    
    entry_weight = Entry(Point(280,100),5)
    entry_weight.setFill('white')
    entry_weight.draw(win)
    
    entry_bmi = Entry(Point(280,130),5)
    entry_bmi.setFill('white')
    entry_bmi.draw(win)
    
    while True:
        click = win.checkMouse()
        if click != None:
            if is_button(click,rect_1):
                rect_1.setFill('grey')
                time.sleep(1)
                rect_1.setFill('yellow')
                height = int(entry_height.getText())/39.37
                weight = int(entry_weight.getText())/2.2
                bmi = round(weight / (height * height), 2)
                entry_bmi.setText(bmi)
            elif is_button(click,rect_2):
                break;
    
    win.close()
    
    
main()
```
