```python
# =============================================================================
# Assignment: 8
# Filename: question5.py
# Description: The Program that creates a GUI window
# Date: 03/02/2020
# Author: Tarini Dash
# =============================================================================

from graphics import *
from random import randrange

# function to check of the click occurred in a circle        
def is_button(click, circle, radius):
    center = circle.getCenter()
    if ((click.getX() - center.getX())**2 + (click.getY() - center.getY())**2 < radius**2):
        return True
    
def getRandomX():
    return randrange(100,300)

def getRandomY():
    return randrange(100,200)

def move(dx,dy,circle):
    circle.move(dx,dy)
    return circle
    
def main():
    win = GraphWin('Question-3', 400, 300)
    p1 = Point(50,50)
    p2 = Point(350,250)
    rect = Rectangle(p1,p2)
    rect.setOutline('black')
    rect.draw(win)
    
    radius = 25
    circle = Circle(Point(getRandomX(),getRandomY()),radius)
    circle.setFill('blue')
    circle.draw(win)
    dx = 1
    dy = 1
    count = 0

    while True:
        time.sleep(5)
        click = win.checkMouse()
        if click != None:
            if(count > 0):
                break
            elif is_button(click,circle,radius):
                circle = move(dx,dy,circle)
                center = circle.getCenter()
                if ((center.getX() - radius) <= 50) or ((center.getX() + radius) >= 350):
                    dx = - dx
                if ((center.getY() - radius) <= 50) or ((center.getY() + radius) >= 250):
                    dy = - dy
        else:
#             Not sure how to disappear the circle . So moved to far away to galaxy
            move(400,400,circle) 
            Text(Point(200,150), "You missed the circle!").draw(win)
            Text(Point(200,160), "Game Over! Click mouse to close the window").draw(win)
            count += 1

                
    win.close()     
          
        
if __name__=="__main__":        
    main()              
```
