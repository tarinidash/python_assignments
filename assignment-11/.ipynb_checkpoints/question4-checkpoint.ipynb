```python
# =============================================================================
# Assignment: 11
# Filename: question4.py
# Description: matplotlib assignments
# Date: 04/04/2020
# Author: Tarini Dash
# =============================================================================

import numpy as np
import matplotlib.pyplot as plt


def main():
    alpha = np.arange(0., 2.1 * 3.14, 0.1)
    r_range = [0.8,1.0,1.2]
    for r in r_range:
       plt.plot((r + np.cos(alpha))*np.cos(alpha), (r + np.cos(alpha))*np.sin(alpha),label = "r="+str(r))

    plt.xlabel('x=r*cos(alpha)')
    plt.ylabel('x=r*sin(alpha)')
    plt.title('Limacon')
    plt.legend(loc='upper right', borderaxespad=0.)
    

#invoke main method
if __name__ == "__main__":
    main()
```
