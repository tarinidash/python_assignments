```python
# =============================================================================
# Assignment: 11
# Filename: question2.py
# Description: matplotlib assignments
# Date: 04/04/2020
# Author: Tarini Dash
# =============================================================================

import pandas as pd
import math
import matplotlib.pyplot as plt


def read(file_location):
    file = pd.ExcelFile(file_location)
    return file.parse("HeightsWeights")
    

def main():
    df = read("children.xlsx")
    plt.scatter(df.Height,df.Weight,color='r')
    plt.xlabel('Heights')
    plt.ylabel('Weights')
    plt.title('Scattered plots, Weights vs. Heights')
    

    
#invoke main method
if __name__ == "__main__":
    main()
```
