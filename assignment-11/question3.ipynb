```python
# =============================================================================
# Assignment: 11
# Filename: question3.py
# Description: matplotlib assignments
# Date: 04/04/2020
# Author: Tarini Dash
# =============================================================================


import numpy as np
import matplotlib.pyplot as plt

def f(x,p):
    return np.exp(-x/p)*np.sin(np.pi*x)
    

def main():
    x = np.arange(0, 10, 0.1)
    p_range = [1,2,5]
    for p in p_range:
       plt.plot(x, f(x,p),label = "p="+str(p))

    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Function f(x) = exp(-x/p)six(pi x),p= 1,2,5')
    plt.legend(bbox_to_anchor=(0.675, 1), loc=2, borderaxespad=0.)

#invoke main method
if __name__ == "__main__":
    main()    
```
