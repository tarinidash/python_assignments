```python
# =============================================================================
# Assignment: 11
# Filename: question5.py
# Description: matplotlib assignments
# Date: 04/04/2020
# Author: Tarini Dash
# =============================================================================


import numpy as np
import matplotlib.pyplot as plt


def main():
    v = 100 
    g = 9.8  
    angle = np.arange(1,8,1)*np.pi/16

    t_max = ((2 * v) * np.sin(angle)) / g
    t = t_max*np.linspace(0,1,100)[:,None]
    for a in angle:
        plt.plot((v * t) * np.cos(angle),(v * t) * np.sin(angle) - g * (t ** 2)/2)
#         plt.plot((v * t) * np.cos(angle),(v * t) * np.sin(angle) - g * (t ** 2)/2,label = str(a))
    
    plt.xlabel('x = V*t*cos(θ) ')
    plt.ylabel('y = V*t*sin(θ) – (gt**2)/2')
    plt.title('Projectile trajectories')
    plt.legend(['1pi/16','2pi/16','3pi/16','4pi/16','5pi/16','6pi/16','7pi/16',])

#invoke main method
if __name__ == "__main__":
    main()
```
