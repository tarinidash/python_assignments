```python
# =============================================================================
# Assignment: 10
# Filename: question2.py
# Description: Numpy assignments
# Date: 03/25/2020
# Author: Tarini Dash
# =============================================================================


import pandas as pd
import math
from pandas_datareader import data as web


def rel_std(stock_data):
    rows = len(stock_data)
    avg = sum(stock_data)/rows
    var = sum(pow(x-avg,2) for x in stock_data) / rows   
    return round(math.sqrt(var)/avg*100,2)
    
    

def main():
    print("Program that calculates relative standard deviation \nfor eight Dow Jones companies.", end = "\n\n")
    
    stocks = {}
    
    ba = web.DataReader("BA", 'yahoo', '1/1/2019', '12/31/2019')
    ba = ba.drop(['Adj Close'], axis = 1)
    ba_close = ba.Close
    ba_std = rel_std(ba_close)
    stocks["Boeing "] = ba_std
    
    ko = web.DataReader("KO", 'yahoo', '1/1/2019', '12/31/2019')
    ko = ko.drop(['Adj Close'], axis = 1)
    ko_close = ko.Close
    ko_std = rel_std(ko_close)
    stocks["Coca Cola"] = ko_std
    
    xom = web.DataReader("XOM", 'yahoo', '1/1/2019', '12/31/2019')
    xom = xom.drop(['Adj Close'], axis = 1)
    xom_close = xom.Close
    xom_std = rel_std(xom_close)
    stocks["Exxon Mobil"] = xom_std
    
    ge = web.DataReader("GE", 'yahoo', '1/1/2019', '12/31/2019')
    ge = ge.drop(['Adj Close'], axis = 1)
    ge_close = ge.Close
    ge_std = rel_std(ge_close)
    stocks["General Electric"] = ge_std
    
    jpm = web.DataReader("JPM", 'yahoo', '1/1/2019', '12/31/2019')
    jpm = jpm.drop(['Adj Close'], axis = 1)
    jpm_close = jpm.Close
    jpm_std = rel_std(jpm_close)
    stocks["JP Morgan"] = jpm_std
    
    nike = web.DataReader("NKE", 'yahoo', '1/1/2019', '12/31/2019')
    nike = nike.drop(['Adj Close'], axis = 1)
    nike_close = nike.Close
    nike_std = rel_std(nike_close)
    stocks["Nike   "] = nike_std
    
    pfe = web.DataReader("PFE", 'yahoo', '1/1/2019', '12/31/2019')
    pfe = pfe.drop(['Adj Close'], axis = 1)
    pfe_close = pfe.Close
    pfe_std = rel_std(pfe_close)
    stocks["Pfeizer"] = pfe_std
    
    vz = web.DataReader("VZ", 'yahoo', '1/1/2019', '31/12/2019')
    vz = vz.drop(['Adj Close'], axis = 1)
    vz_close = vz.Close
    vz_std = rel_std(vz_close)
    stocks["Verizon"] = vz_std
    
    
    print("Company \t\t Relative std deviation")
    for key in stocks:
        print(key, "\t\t" ,stocks[key])

    
#invoke main method
if __name__ == "__main__":
    main()
```
